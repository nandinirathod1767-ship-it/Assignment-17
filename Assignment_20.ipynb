```python
filepath='/content/ford_car_dataset.csv'  # Load the Ford Car Dataset
df=pd.read_csv(filepath)
```

```python
df = pd.read_csv("ford_car_dataset.csv")

X = df.drop("price", axis=1)
Y = df["price"]

print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
categorical_columns = X.select_dtypes(include=['object']).columns
print("Categorical Columns:")
print(categorical_columns)

X_encoded = pd.get_dummies(X, columns=categorical_columns)

X_encoded = X_encoded.astype(int)

# Show the first 5 rows
print("\nFirst 5 Rows After One-Hot Encoding:")
print(X_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  # Create StandardScaler object
num_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
X_encoded[num_cols] = scaler.fit_transform(X_encoded[num_cols])
print("First 5 Rows Feature Scaling:")
print(X_encoded.head())
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    Y,
    test_size=0.33,
    random_state=42
)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)


print("Intercept:", model.intercept_)

print("\nCoefficients:")
print(model.coef_)
```

```python

y_pred = model.predict(X_test)

print("First 10 Predicted Values:")
print(y_pred[:10])

print("\nFirst 10 Actual Values:")
print(y_test.head(10))
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
```

```python
import joblib

joblib.dump(model, "LR_ford_car.pkl")

joblib.dump(scaler, "scaler.pkl")

joblib.dump(X_encoded.columns.tolist(), "columns.pkl")

print("Model, Scaler, and Columns saved successfully!")
```
