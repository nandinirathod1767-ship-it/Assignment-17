```python
# Sex
sns.countplot(x="sex", data=df)
plt.title("Sex Count")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()

# Smoker
sns.countplot(x="smoker", data=df)
plt.title("Smoker Count")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()

# Region
sns.countplot(x="region", data=df)
plt.title("Region Count")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```
